```python
# We add liquidity factor as Bid-Ask spread
liquidity = pd.read_excel("./data_files/liquidity.xlsx")

print(liquidity)
```

```python
liquidity_melted = liquidity.melt(id_vars="date", value_vars=liquidity[1:529], var_name="isin", value_name="liquidity")
print(liquidity_melted)
```

```python
liquidity_melted_clean = liquidity_melted.dropna()

print(liquidity_melted_clean)
```

```python
liquidity_melted_clean['isin'] = pd.factorize(liquidity_melted_clean['isin'])[0]
```

```python
working_df_clean = working_df_clean.merge(liquidity_melted_clean, on=["date","isin"], how="left")
print(working_df_clean)
```

```python
print(working_df_clean.head())
```

```python
working_df_clean['rating'] = working_df_clean['rating'].str.strip()
```

```python
rating_map = {
    'AAA': 1, 'AA+': 2, 'AA': 3, 'AA-': 4,
    'A+': 5, 'A': 6, 'A-': 7,
    'BBB+': 8, 'BBB': 9, 'BBB-': 10,
    'BB+': 11, 'BB': 12, 'BB-': 13,
    'B-': 14, 'NR':15
}

working_df_clean['credit_rating_numeric'] = working_df_clean['rating'].map(rating_map)
```

```python
def rating_class(x):
    if x == 1:
        return 'AAA class bonds'
    elif x <= 4:
        return 'AA class bonds'
    elif x <= 7:
        return 'A class bonds'
    elif x <= 14:
        return 'BBB/BB/B bonds'
    else:
        return 'Not rated bonds'
```

```python
working_df_clean['rating_class'] = working_df_clean['credit_rating_numeric'].apply(rating_class)
```

```python
working_df_clean.isna().sum()
```

```python
def maturity_bucket(x):
    if x <= 4:
        return 'Short-Term'
    elif x <= 10:
        return 'Medium-Term'
    else:
        return 'Long-Term'

working_df_clean['maturity_bucket'] = working_df_clean['maturity'].apply(maturity_bucket)
```
