```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
df = pd.read_csv("public.csv")
print(df.head())
```

```python
df= df.drop(columns=['Unnamed: 0', 'type_of_firm'])
print(df)
```

```python
# Assume your full firm-level dataset is called `firm_df`
df['type_of_firm'] = df['stock_return'].apply(
    lambda x: 'Public' if pd.notnull(x) else 'Private'
)

print(df.head())
```

```python
df_long_term = df[df['maturity_bucket'] == "Long-Term"]
print(df_long_term)
```

```python
df_medium_term = df[df['maturity_bucket'] == "Medium-Term"]
print(df_medium_term)
```

```python
results_by_firm_type = {}
r_squared_table = {}

coef_table = pd.DataFrame()
for group_name, group_data in df.groupby('type_of_firm'):
    y_group = group_data['cs_change']
    x_group = group_data[['delta lr', 'change_in_10y_bond', 'change_in_slope', 'smi_return', 'v3x_change', 'cpi_change', 'unempl_change', 'gdp_change','liquidity']]
    x_group = sm.add_constant(x_group)
    model1 = sm.OLS(y_group, x_group).fit()
    
    coef = model1.params
    coef.name = group_name

    coef_table = pd.concat([coef_table, coef], axis=1)
    
    print(f"\n\nResult for {group_name}")
    print(model1.summary)
    print(model1.summary())
    results_by_firm_type[group_name] = model1
    
    r_squared_table[group_name] = model1.rsquared
    

coef_table = coef_table.T
```

```python
r_squared_table = pd.Series(r_squared_table, name='R_squared')
print(r_squared_table)
```

### Fixed effect regression

```python
panel_data_adj = df.set_index(['date', 'isin']).sort_index()
print(panel_data_adj.head())
```

```python
from linearmodels.panel import PanelOLS
```

```python
results_by_firm_type = {}
r_squared_table = {}

coef_table = pd.DataFrame()
for group_name, group_data in panel_data_adj.groupby('type_of_firm'):
    y_group = group_data['cs_change']
    x_group = group_data[['delta lr', 'change_in_10y_bond', 'change_in_slope', 'smi_return', 'v3x_change', 'cpi_change', 'unempl_change', 'gdp_change', 'liquidity']]
    x_group = sm.add_constant(x_group)  
    model = PanelOLS(y_group, x_group, entity_effects=True, drop_absorbed=True).fit()
    
    coef = model.params
    coef.name = group_name
    coef_table = pd.concat([coef_table, coef], axis=1)
    
    
    print(f"\n\nResult for {group_name}")
    print(model.summary)
    
  
    results_by_firm_type[group_name] = model
    r_squared_table[group_name] = model.rsquared
    

coef_table = coef_table.T
```

```python
# Create R² DataFrame
r_squared_table = pd.Series(r_squared_table, name='R_squared')
# Print
print(r_squared_table)
```
